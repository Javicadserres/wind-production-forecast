```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import cm
from pathlib import Path
```

#### Import data

```python
PATH_DATA = Path.cwd().parent / 'data' / 'files'

data = pd.read_csv((PATH_DATA / 'data.csv'), index_col='time')
```

#### PLOTS

```python
plt.figure(figsize=(8, 4))
plt.scatter(data['u100_mean_36.0-5.45'].abs(), data['production'], s=5)
plt.xlabel('Velocidad')
plt.ylabel('Producción energía eólica')
plt.title('Relación entre la velocidad del viento y la producción de energía en Cádiz')
plt.grid()
```
